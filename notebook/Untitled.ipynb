```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('propulsion.csv')
```

```python
data.drop('Unnamed: 0', inplace=True,axis=1)
```

```python
corr_mat = data.corr()
```

```python
fig = plt.figure(figsize=(20,15))
plt.style.use('fast')


sns.heatmap(corr_mat, annot=True,cmap="Blues",linewidth=0.3, vmin= 0, vmax=1, cbar_kws={"shrink": .8});

plt.tick_params(axis='x',labelsize=10)
plt.tick_params(axis='y', labelsize=15)
plt.show()
```

```python
mask = np.triu(np.ones_like(corr_mat, dtype=np.bool))
```

```python
mask.shape
```

```python
corr_mat.shape
```

```python
mask[1:, :-1].shape
```

```python
# adjust mask and df
mask = mask[1:, :-1]
corr = corr_mat.iloc[1:,:-1].copy()
# plot heatmap
fig = plt.figure(figsize=(20,15))

sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='Blues',vmin=-1, vmax=1, cbar_kws={"shrink": .8})
# yticks
plt.yticks(rotation=0)
plt.show()
```
